```python
# 数据的时间分布
# 使用apply()函数和lambda表达式提取'date'列中的年份和月份，生成新列'new_date'
df['new_date'] = df['date'].apply(lambda seq: seq[:10])  # 只提取这一列数据的前10位

# 对不同时间的帖子数量进行计数
count_per_day = df['new_date'].value_counts()

# 按时间顺序对帖子数量进行排序
count_per_day_sorted = dict(sorted(count_per_day.items(), key=lambda item: item[0]))  # 时间排序

# 将时间字符串转换为datetime对象，并创建numpy数组
x = np.array([datetime.strptime(k, '%Y-%m-%d') for k in count_per_day_sorted.keys()])
y = np.array(list(count_per_day_sorted.values()))

# 绘制帖子数量随时间变化的折线图
f = plt.figure(figsize=(14, 6))
plt.plot(x, y)
#plt.title('Number of posts per day')
plt.xlabel('Date')
plt.ylabel('Number of posts')
plt.grid()
plt.show()

# 保存折线图为PDF文件
f.savefig("Number of posts per day.pdf")
```
