```python
df['new_date'] = df['date'].apply(lambda seq: seq[:10])  # Âè™ÊèêÂèñËøô‰∏ÄÂàóÊï∞ÊçÆÁöÑÂâç10‰Ωç
df['new_date']
```

```python
count_per_day = df['new_date'].value_counts()
count_per_day
```

```python
count_per_day_sorted = dict(sorted(count_per_day.items(), key=lambda item: item[0]))  # Êó∂Èó¥ÊéíÂ∫è
count_per_day_sorted
```

```python
# Êï∞ÊçÆÁöÑÊó∂Èó¥ÂàÜÂ∏É
# ‰ΩøÁî®apply()ÂáΩÊï∞ÂíålambdaË°®ËææÂºèÊèêÂèñ'date'Âàó‰∏≠ÁöÑÂπ¥‰ªΩÂíåÊúà‰ªΩÔºåÁîüÊàêÊñ∞Âàó'new_date'
df['new_date'] = df['date'].apply(lambda seq: seq[:10])  # Âè™ÊèêÂèñËøô‰∏ÄÂàóÊï∞ÊçÆÁöÑÂâç10‰Ωç

# ÂØπ‰∏çÂêåÊó∂Èó¥ÁöÑÂ∏ñÂ≠êÊï∞ÈáèËøõË°åËÆ°Êï∞
count_per_day = df['new_date'].value_counts()

# ÊåâÊó∂Èó¥È°∫Â∫èÂØπÂ∏ñÂ≠êÊï∞ÈáèËøõË°åÊéíÂ∫è
count_per_day_sorted = dict(sorted(count_per_day.items(), key=lambda item: item[0]))  # Êó∂Èó¥ÊéíÂ∫è

# Â∞ÜÊó∂Èó¥Â≠óÁ¨¶‰∏≤ËΩ¨Êç¢‰∏∫datetimeÂØπË±°ÔºåÂπ∂ÂàõÂª∫numpyÊï∞ÁªÑ
x = np.array([datetime.strptime(k, '%Y-%m-%d') for k in count_per_day_sorted.keys()])
y = np.array(list(count_per_day_sorted.values()))

# ÁªòÂà∂Â∏ñÂ≠êÊï∞ÈáèÈöèÊó∂Èó¥ÂèòÂåñÁöÑÊäòÁ∫øÂõæ
f = plt.figure(figsize=(14, 6))
plt.plot(x, y)
#plt.title('Number of posts per day')
plt.xlabel('Date')
plt.ylabel('Number of posts')
plt.grid()
plt.show()

# ‰øùÂ≠òÊäòÁ∫øÂõæ‰∏∫PDFÊñá‰ª∂
f.savefig("Number of posts per day.pdf")
```
